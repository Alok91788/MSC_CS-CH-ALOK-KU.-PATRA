openCV Basics

*   Convert to grayscale
*   resize the image
*   Draw shapes (rectangle,text)
*   save the modify image





```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow # Removed extra dot from this line

from google.colab import files
uploaded = files.upload()
```

```python
#step 3: Read the upload image
filename=list(uploaded.keys())[0]
image=cv2.imread(filename)#openCV reads in BGR format
cv2_imshow(image)
```

```python
#step 4:convert to Grayscale
gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

```python
#step 5:Resize the image to 300x300
resized=cv2.resize(gray,(300,300))
cv2_imshow(resized)
#step 5:Resize the image to 300x300
resized=cv2.resize(image,(300,300))
cv2_imshow(resized)
```

```python
#step 5:Resize the image to 300x300
a=[gray , image]
for i in a:
  resized=cv2.resize(i,(150,150))
  cv2_imshow(resized)
```

```python
#step 6: Draw Rectangle and Text
annoted=image.copy()
an=cv2.resize(an,(150,150))
cv2.rectangle(an,(50,50),(100,100),(0,255,0),2)
cv2.putText(an,'My Box',(50,35),cv2.FONT_HERSHEY_SIMPLEX,1,(255,0,0),2)
cv2_imshow(an)
```

```python
#step 7:Save the modify image
cv2.imwrite('output_image.jpg',annoted)
print("Imagw saved as output_image,jpg")
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator,load_img, img_to_array
```

```python
from google.colab import files
uploaded = files.upload()
filename=list(uploaded.keys())[0]
```

```python
img=load_img(filename,target_size=[224,224])
img_array=img_to_array(img)
img_array=np.expand_dims(img_array,axis=0)
```

```python
datagen=ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    brightness_range=[0.5,1.5]
)
```

```python
i=0
for batch in datagen.flow(img_array,batch_size=1):
  plt.figure(figsize=(2,2))
  plt.imshow(batch[0].astype('uint8')) # Changed 'unit8' to 'uint8'
  plt.axis('off')
  i+=1
  if i>5:
    break
```
